# Text Summarization using DistilBART

This Code Template is for Text Summarization task one of the problem associated to NLP(Natural Language Processing). In this Code template DistilBART is utilized, which is transformer based model.

```python
from transformers import pipeline
import warnings
warnings.filterwarnings("ignore")
```

### DistilBART



```python
summarizer = pipeline("summarization", model="philschmid/distilbart-cnn-12-6-samsum")
```




### Text Summarization

```python
conversation = '''Jeff: Can I train a 🤗 Transformers model on Amazon SageMaker? 
Philipp: Sure you can use the new Hugging Face Deep Learning Container. 
Jeff: ok.
Jeff: and how can I get started? 
Jeff: where can I find documentation? 
Philipp: ok, ok you can find everything here.                                         
'''
summarizer(conversation,max_length=30)
```
