## Car Number Plate Detection using OpenCV

This code template is for car number plate detection using OpenCV. This task uses the Haar Cascade XML file containing the features of Russian car plates for detecting the bounding box for the license plates. 

## Required Packages

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Importing our Classifier

**Link:** [classifier used in this template](https://drive.google.com/file/d/15o6VctOvVgySRoQMigPIRly9oBnU4Q1L/view)

```python
cascade = cv2.CascadeClassifier('')
```

## Reading our image

OpenCV-Python is a library of Python bindings designed to solve computer vision problems.

We use the the **"cv2.imread()"** method to load an image from the specified file path.

```python
cars = cv2.imread('')
```

## Converting Image to Gray

```python
gray = cv2.cvtColor(cars, cv2.COLOR_BGR2GRAY)
```

## Converting image to colored image

```python
def converToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

## Drawing bounding boxes

```python
cars_detected = cascade.detectMultiScale(
        gray,
        scaleFactor=1.2,
        minNeighbors=5,
        minSize=(20,20))
```

```python
print('Found Numbers', len(cars_detected))
```

```python
for (x, y, w, h) in cars_detected:
    cv2.rectangle(cars, (x,y), (x+w,y+h),(145,60,255), 5)
```

## Final Image 

```python
plt.figure(figsize=(20,20))
plt.imshow(converToRGB(cars));
```
